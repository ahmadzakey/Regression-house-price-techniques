# Handle Missing value

```python
df_cleaned = df_1.copy()
```

```python
# Mengisi missing value dalam lajur numerik dengan median
df_cleaned['LotFrontage'].fillna(df_cleaned['LotFrontage'].median(), inplace=True)
df_cleaned['MasVnrArea'].fillna(df_cleaned['MasVnrArea'].median(), inplace=True)
df_cleaned['GarageYrBlt'].fillna(df_cleaned['GarageYrBlt'].median(), inplace=True)

```

```python
# Mengisi missing dalam lajur categorical dengan mode
for column in ['Alley', 'MasVnrType', 'BsmtQual', 'BsmtExposure', 'BsmtFinType1', 
                 'BsmtFinType2', 'BsmtCond', 'GarageType', 'GarageFinish', 
                 'GarageQual', 'GarageCond', 'PoolQC', 'Fence', 'MiscFeature','FireplaceQu','Electrical']:
    if df_cleaned[column].isnull().sum() / len(df_cleaned) > 0.1:
        # Jika missing value signifikan, gantikan dengan 'Unknown' ( melebihi 10%)
        df_cleaned[column].fillna('Unknown', inplace=True)
    else:
        # Jika missing value kecil, gunakan mode
        df_cleaned[column].fillna(df_cleaned[column].mode()[0], inplace=True)
```
